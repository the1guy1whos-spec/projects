```python
#wordle


import random
import sys
from nltk.corpus import words

def check_5():
    word = ""
    while len(word) != 5:
        word = random.choice(words.words())
    return word


def list_(word):
    word_list = list(word)
    return word_list

def first_letter(the_word, player_guess):

    if the_word[0] == player_guess[0]:
        one = 'correct'
        return one
    elif the_word[1]==player_guess[0] or the_word[2]==player_guess[0] or the_word[3]==player_guess[0] or the_word[4]==player_guess[0]:
        one = 'displaced'
        return one
    else:
        one = 'not present'
        return one

def second_letter(the_word, player_guess):

    if the_word[1] == player_guess[1]:
        one = 'correct'
        return one
    elif the_word[0]==player_guess[1] or the_word[2]==player_guess[1] or the_word[3]==player_guess[1] or the_word[4]==player_guess[1]:
        one = 'displaced'
        return one
    else:
        one = 'not present'
        return one
    
def third_letter(the_word, player_guess):

    if the_word[2] == player_guess[2]:
        one = 'correct'
        return one
    elif the_word[0]==player_guess[2] or the_word[1]==player_guess[2] or the_word[3]==player_guess[2] or the_word[4]==player_guess[2]:
        one = 'displaced'
        return one
    else:
        one = 'not present'
        return one
    

def forth_letter(the_word, player_guess):

    if the_word[3] == player_guess[3]:
        one = 'correct'
        return one
    elif the_word[0]==player_guess[3] or the_word[1]==player_guess[3] or the_word[2]==player_guess[3] or the_word[4]==player_guess[3]:
        one = 'displaced'
        return one
    else:
        one = 'not present'
        return one
    
def fifth_letter(the_word, player_guess):

    if the_word[4] == player_guess[4]:
        one = 'correct'
        return one
    elif the_word[0]==player_guess[4] or the_word[1]==player_guess[4] or the_word[2]==player_guess[4] or the_word[3]==player_guess[4]:
        one = 'displaced'
        return one
    else:
        one = 'not present'
        return one
    
def length(player_guess, real_word):
    if len(player_guess)!= 5:
        sys.exit()
    elif player_guess not in real_word:
        sys.exit()




def main():
    correct_answer=list_(check_5())
    print(correct_answer)
    list_guess = []
    guess_number = 0
    times = 0
    while times < 4 or list_guess != correct_answer:
        guess_number += 1
        if guess_number == 1:
            number_ending = 'st'
        elif guess_number == 2:
            number_ending = 'nd'
        elif guess_number == 3:
            number_ending = 'rd'
        elif guess_number == 4 or guess_number == 5:
            number_ending = 'th'

        if times == 5:
            sys.exit()
        guess = input(f"what is your {guess_number}{number_ending} guess")
        sub = words.words()
        length(guess, sub)
        list_guess = list(guess)
        print(list_guess)
        print(first_letter(correct_answer, list_guess))
        print(second_letter(correct_answer, list_guess))
        print(third_letter(correct_answer, list_guess))
        print(forth_letter(correct_answer, list_guess))
        print(fifth_letter(correct_answer, list_guess))
        times += 1
        print(times)
    main()
```
