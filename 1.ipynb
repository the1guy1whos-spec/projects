```python
#convertion calc

import sys as sy
def in_start(original_value):
    convert_inches = input ("What unit would you like to convert from inches?")
    if convert_inches == "cm" or convert_inches == "centimeter":
        inches_conversion = float(original_value)*2.54
        return "cm", inches_conversion
    elif convert_inches == "mm" or convert_inches == "millimeter":
        inches_conversion = float(original_value)*25.4
        return "mm", inches_conversion
    elif convert_inches == "meter" or convert_inches =="m":
        inches_conversion = float(original_value)*.0254
        return "m", inches_conversion

def cm_start(original_value):
    convert_cm = input ("What unit would you like to convert from cm?")
    if convert_cm == "inch" or convert_cm == 'in':
        cm_conversion = float(original_value)/2.54
        #converts to inches
        return "in", cm_conversion
    elif convert_cm == "mm" or convert_cm == "millimeter":
        cm_conversion = float(original_value)*10
        #converts to millimeter
        return "mm", cm_conversion
    elif convert_cm == "meter" or convert_cm == "m":
        cm_conversion = float(original_value)/100
        #converts to meter
        return "m", cm_conversion

def km_start(original_value):
    convert_km = input ("What unit would you like to convert from km?")
    if convert_km == "feet" or convert_km == "ft":
        km_conversion = float(original_value)*3280.84
        return "ft", km_conversion
    elif convert_km == "miles" or convert_km == "mi":
        km_conversion = float(original_value)*0.621371
        return "mi", km_conversion
    elif convert_km == "inch" or convert_km == "in":
        km_conversion = float(original_value)*39370.1
        return "in", km_conversion

def mi_start(original_value):
    convert_mi = input ("What unit would you like to convert from mi?")
    if convert_mi == "feet" or convert_mi == "ft":
        mi_conversion = float(original_value)*5280
        return "ft", mi_conversion
    elif convert_mi == "kilometer" or convert_mi == "km":
        mi_conversion = float(original_value)*1.60934
        return "km", mi_conversion
    elif convert_mi == "inch" or convert_mi == "in":
        mi_conversion = float(original_value)*63360
        return "in", mi_conversion

def main():
    unit = input("What is your starting unit? (mi, km, in, cm)")
    if unit not in ['mi', 'km', 'in', 'cm']:
        print('Sorry, I have not added that unit yet.')
        sy.exit()

    value = float(input("What is the value that you would like to convert?"))
    if unit == 'mi':
        unit_2, final_conversion = mi_start(value)
        print(f"The converted value is about {final_conversion:.2f} {unit_2}.")
    elif unit == 'in':
        unit_2, final_conversion = in_start(value)
        print(f"The converted value is about {final_conversion:.2f} {unit_2}.")
    elif unit == 'cm':
        unit_2, final_conversion = cm_start(value)
        print(f"The converted value is about {final_conversion:.2f} {unit_2}.")
    elif unit == 'km':
        unit_2, final_conversion = km_start(value)
        print(f"The converted value is about {final_conversion:.2f} {unit_2}.")


main()
```
