```python
#guess number game


import random as ran
def number_choosing():
    number = ran.randint(1,100)
    return number

def big_small(guess, number):
    if guess < number:
        determination = "go bigger"
        return determination
    elif guess > number:
        determination = "go smaller"
        return determination

while True:
    number = number_choosing()
    guess = int(input('What is your guess?'))
    times = 1
    while guess != number and times < 10:
        determination = big_small(guess, number)
        guess = int(input(f'Try again, this time {determination}.'))
        times += 1
    if guess == number:
        again = input('You got it right:) Play again?(Yes/No)')
        if again == 'Yes' or again =='yes':
            print('starting again')
        else:
            break
    else:
        again = input('You got it wrong:( Play again?(Yes/No)')
        if again == 'Yes' or again =='yes':
            print('starting again')
        else:
            break
```
