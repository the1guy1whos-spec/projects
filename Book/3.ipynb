```python
#Rock, Paper, Scissor Shoot

import random as ran
choices = ['rock', 'scissor', 'paper']
def comp_choice(choices):
    computer = ran.choice(choices)
    return computer


def check_winner(comp, player):
    if comp == player:
        result = 'tied with'
        reaction = ': |'
        return result, reaction
    elif comp == 'rock' and player == 'paper' or comp == 'paper' and player == 'scissor' or comp == 'scissor' and player == 'rock':
        result = 'won against'
        reaction = ':)'
        return result, reaction
    elif comp == 'rock' and player == 'scissor' or comp == 'paper' and player == 'rock' or comp == 'scissor' and player == 'paper':
        result = 'lost to'
        reaction = ':('
        return result, reaction
def check_the_choice(player_, choices):
    if player_ not in choices:
        check_choice = "no"
        return check_choice
    else:
        pass
while True:
    computer = comp_choice(choices)
    player = input ('Choose between Rock, Paper, or Scissor')
    player_ = player.lower()
    check_choice = check_the_choice(player_, choices)
    if check_choice == 'no':
        input('Invalid Choice!!! Try Again >:(')
    result, reaction = check_winner(computer, player_)
    again = input(f'You {result} computer {reaction} Play again? Yes/No')
    if again == 'Yes' or again == 'yes':
        print('starting again')
    else:
        break
```
